### Earthquake Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('F:\python project\practice\Data\earthquake.csv')


Data = {
     'Location': df['Location'].values,
     'Magnitude': df['Magnitude'].values,
     'Country_Code': ['NPL', 'ECU', 'MEX', 'IDN', 'USA', 'NZL', 'GRC', 'HTI', 'JPN', 'TUR']
}


Earth_data = pd.DataFrame(Data)
# Earth_data.drop(9, inplace= True)
print(df.head())
print(Earth_data.info())
```

### plotting with customization

```python
colors = plt.cm.viridis(np.linspace(0, 1, len(df)))


plt.figure(figsize=(10,6))
plt.bar(df['Location'], df['Magnitude'], color=colors)
plt.xlabel('Location')
plt.ylabel('Magnitude')
plt.title('Earthquake Magnitudes by Location')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

### chloropeth Map for this Data

```python
import plotly.express as px

fig = px.choropleth(
    Data,
    locations='Country_Code',
    color='Magnitude',
    hover_name='Location',
    color_continuous_scale=px.colors.sequential.Plasma,
    title='Earthquake Magnitudes by Location',
    labels={'Magnitude': 'Magnitude'},
    projection='natural earth'
)

# Update layout for better visualization
fig.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=True,
        projection_type='equirectangular'
    ),
    margin={"r": 0, "t": 40, "l": 0, "b": 0}
)

# Show the map
fig.show()
```
