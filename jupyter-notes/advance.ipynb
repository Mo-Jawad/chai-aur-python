### Skewness and Kurtosis

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

data = [12, 15, 18, 19, 20, 22, 22, 24, 30, 40, 60, 33, 35, 37, 40, 42]

data_series = pd.Series(data)

#measuring skewness and kurtosis
skewness = data_series.skew()

kurtosis = data_series.kurtosis()

print(f"Skewness: {skewness}")
print(f"kurtosis: {kurtosis}")

# determine figure size
plt.figure(figsize=(10,6))

sns.histplot(data_series, kde=True, bins=10, color="skyblue", stat="density")

plt.title("Distribution of Data with KDE", fontsize=16)
plt.xlabel("Data Values", fontsize=12)
plt.ylabel("Density", fontsize=12)

plt.show()
```
